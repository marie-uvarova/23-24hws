#### Задача 1 (2 балла).

Реализуйте с помощью классов базу данных для ветеринарного магазина из предыдущего семинара. Видимо, нам понадобятся два класса: для владельцев животных и для самих животных. У владельца мы храним имя, фамилию и номер телефона. У животного пусть будет кличка, возраст и дополнительный атрибут "записан на прием", в котором будет храниться информация о дате следующего приема (если ее нет - None). База данных должна храниться в бинарном файле (используйте pickle или dill). При запуске программы должны быть доступны несколько действий: завести нового клиента, завести новое животное для существующего клиента и записать животное на прием (или отметить, что прием прошел и новой записи нет). Сами экземпляры клиентов хранятся просто в списке.

Примечание: в атрибутах могут храниться не только строки, но и списки, но не забывайте об их особенностях. 

```python
import pickle

class Owner:
     def __init__(self, name, surname, phone_number):
          self.name = name
          self.surname = surname
          self.phone_number = phone_number
     
     def __str__(self):
        return f'{self.name} {self.surname}, {self.phone_number}'
        
     def __repr__(self):
        return f"Owner('{self.name}, {self.surname}, {self.phone_number}')"

class Pet:
     def __init__(self, pet_name, age, appointment):
          self.pet_name = pet_name
          self.age = age
          self.appointment = appointment

     def __str__(self):
          return f'{self.pet_name}, {self.age}, {self.appointment}'
        
     def __repr__(self):
        return f"Pet('{self.pet_name}, {self.age}, {self.appointment}')"

     def get_app(self, date, status):
          self.appointment = date + ', ' + status
          print(self.appointment)

clients = []
dct = dict()
pickle.dump(dct, open('database', 'wb'))

print('Enter 1 to add a new client.')
print('Enter 2 to add a pet to an existing client.')
print('Enter 3 to get an appointment')
print('Enter 0 to stop.')
x = int(input())
while x != 0:
     if x == 1:
          nm = input("Insert a client's name: ")
          lastnm = input("Insert a client's surname: ")
          phone = input("Insert a client's phone number: ")
          client = Owner(nm, lastnm, phone)
          clients.append(client)

          print(Owner.__str__(client))

     if x == 2:
          ownersnm = input("Insert the owner's name: ")
          petnm = input("Insert a pet's name: ")
          age = input("Insert a pet's age: ")
          appointment = None
          pet = Pet(petnm, age, appointment)
          if ownersnm not in dct:
               dct[ownersnm] = [pet]
          else:
               dct[ownersnm].append(pet)

          print(Pet.__str__(pet))

     if x == 3:
          ownersnm = input("Insert the owner's name: ")
          a = dct.get(ownersnm)
          petnm = input("Insert the pet's name: ")
          for elem in a:
               elem2 = Pet.__str__(elem)
               s = elem2.split(', ')
               print(ownersnm, ' ', s[0])
               if s[0] == petnm:
                    date = input("Insert the appointment's date and time: ")
                    stts = input("Insert the appointment's status: ")
                    Pet.get_app(elem, date, stts)

     pickle.dump(dct, open('database', 'wb'))
     x = int(input())


data_bn = pickle.load(open('database', 'rb'))
output = []
for elem in data_bn:
     a = data_bn.get(elem)
     for elem2 in a:
          elem3 = Pet.__str__(elem2)
          output.append(elem3)
     print(f'{elem}: {str(output).strip("[]")}')
```
