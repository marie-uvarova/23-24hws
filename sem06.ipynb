#### Задача 1. 

Реализуйте для класса Human метод \_\_bool\_\_, который будет проверять нашего человека на совершеннолетие. 

```python
class Human:
    def __init__(self, name, surname, age):
        self.name = name
        self.surname = surname
        self.age = age

    def __str__(self):
        return f'{self.name} {self.surname}, {self.age}'
    
    def __repr__(self):
        return f"Human('{self.name}', '{self.surname}', {self.age})"

    def __bool__(self):
        if self.age >= 18:
            print('Adult')
        else:
            print('Underage')

nm = input()
srnm = input()
nm_age = int(input())
person = Human(nm, srnm, nm_age)
print(Human.__str__(person))
Human.__bool__(person)
```

#### Задача 2. 

Напишите класс Linguist, у которого при создании будут повляться четыре динамических атрибута: имя, аффилиация, показатель выносливости (stamina) и список публикаций (по умолчанию пустой). 

```python
from time import sleep # sleep(n) - программа будет ждать n секунд

class Linguist:
    def __init__(self, fullname, affiliation, stamina, publication):
        self.fullname = fullname
        self.affiliation = affiliation
        self.stamina = stamina
        self.publication = publication
        
    def __str__(self):
        return f'{self.fullname}, {self.affiliation}, {self.stamina}'
        
    def __repr__(self):
        return f"Linguist('{self.fullname}, {self.affiliation}, {self.stamina}')"
        
    def work(self, publication_name):
        '''
        Лингвист работает над публикацией n времени (можете придумать, как вычислять его на основе названия публикации).
        За это время он устает в k*n раз (вычитаем это из его выносливости, только следите, чтобы усталость отрицательной не сделалась).
        '''
        n = int(input())
        k = 10
        coef = len(publication_name.split())
        n *= (1 + coef / 100)
        if self.stamina >= n * k:
            self.stamina -= n * k
        else:
            self.stamina = 0

        self.publication.append(publication_name)
        print(f'Сил осталось: {self.stamina}')
    
    def sleep(self, hours):
        '''Лингвист спит n часов, восстанавливая k*n выносливости'''
        sleep(hours)
        k = 10
        if self.stamina < 150:
            self.stamina += hours * k
        else:
            self.stamina = 150
        print(f'Сил осталось: {self.stamina}')
            
        
    def show_publications(self):
        print('Список работ:')
        if len(self.publication) == 0:
            print('У вас ни одной публикации! Быстро работать.')
        for elem in self.publication:
            print(elem)
       
fullnm = input()
affiliate = input()
stamina = int(input())
publication = []

person = Linguist(fullnm, affiliate, stamina, publication)

print(Linguist.__str__(person))

Linguist.show_publications(person)

publ_name = input('Введите название публикации: ')
Linguist.work(person, publ_name)

Linguist.show_publications(person)

n = int(input())
Linguist.sleep(person, n)

publ_name = input('Введите название публикации:')
Linguist.work(person, publ_name)

Linguist.show_publications(person)
```

#### Задача 3. 

Доработаем задачу 2: добавьте класс "публикация", чтобы у публикации была еще некоторая информация о ее объеме (и, соответственно, чтобы можно было вычислять количество времени, которое на нее нужно потратить). А у лингвиста теперь должен появиться показатель "наработал в экспедициях", который будет пополняться в специальном методе go_to_expedition(), и если его значение будет недостаточным для выбранной публикации, нужно об этом предупреждать и публикацию не добавлять. 

```python
from time import sleep

class Linguist:
    
    def __init__(self, fullname, affiliation, stamina, publication):
        self.fullname = fullname
        self.affiliation = affiliation
        self.stamina = stamina
        self.publication = publication
        
    def __str__(self):
        return f'{self.fullname}, {self.affiliation}, {self.stamina}'
        
    def __repr__(self):
        return f"Linguist('{self.fullname}, {self.affiliation}, {self.stamina}')"
        
    def work(self, publication, exp):

        if exp >= 18:
            k = 10
            if self.stamina >= publication.volume * (1 + publication.data / 100) * k:
                self.stamina -= publication.volume * (1 + publication.data / 100) * k
            else:
                self.stamina = 0

            self.publication.append(publication.name)
            print(f'Сил осталось: {self.stamina}')
        else:
            return False
    
    def sleep(self, hours):
        '''Лингвист спит n часов, восстанавливая k*n выносливости'''
        sleep(hours)
        k = 10
        if self.stamina < 150:
            self.stamina += hours * k
        else:
            self.stamina = 150
        print(f'Сил осталось: {self.stamina}')
            
        
    def show_publications(self):
        print('Список работ:')
        if len(self.publication) == 0:
            print('У вас ни одной публикации! Быстро работать.')
        for elem in self.publication:
            print(elem)
    
    def go_to_expedition(self, area, days):
        '''
        Можно придумать что-нибудь и для региона, куда едет лингвист:
        допустим, на Кавказе показатель будет расти быстрее, чем в Поволжье и т.п.
        '''
        exp_areas = {'Поволжье': 1.1,'Карелия': 1.2, 'Кавказ': 1.4, 'Камчатка': 1.4}
        exp_coef = exp_areas.get(area)
        return days * exp_coef
    

class Publication:
    def __init__(self, name, volume, data):
        '''
        name - имя публикации
        volume - количество страниц
        data - количество необходимого материала (в попугаях...)
        '''
        self.name = name
        self.volume = volume
        self.data = data
        
    def __str__(self):
        return f'{self.name}, {self.volume}, {self.data}'
        
    def __repr__(self):
        return f"Publication('{self.name}, {self.volume}, {self.data}')"


fullnm = input()
affiliate = input()
stamina = int(input())
publication = []
person = Linguist(fullnm, affiliate, stamina, publication)

print(Linguist.__str__(person))

Linguist.show_publications(person)

area = input('Введите регион экспедиции: ')
days = int(input('Введите продолжительность экспедиции: '))
exp = Linguist.go_to_expedition(person, area, days)

publ_name = input('Введите название публикации: ')
publ_vol = int(input('Введите количество страниц: '))
publ_data = int(input('Введите количество отглоссированных предложений: '))
publication = Publication(publ_name, publ_vol, publ_data)

print(Publication.__str__(publication))

Linguist.work(person, publication, exp)

if not Linguist.work(person, publication, exp):
    print('Вашего экспедиционного опыта недостаточно. Публикация не может быть добавлена.')
    print('Попробовать еще раз?')
    x = input()
    if x == 'Да':
        area = input('Введите регион экспедиции: ')
        days = int(input('Введите продолжительность экспедиции: '))
        exp = Linguist.go_to_expedition(person, area, days)

        publ_name = input('Введите название публикации: ')
        publ_vol = int(input('Введите количество страниц: '))
        publ_data = int(input('Введите количество отглоссированных предложений: '))
        publication = Publication(publ_name, publ_vol, publ_data)

        print(Publication.__str__(publication))

        Linguist.work(person, publication, exp)
        Linguist.show_publications(person)
    else:
        print('Всего доброго.')
    
```
